```python
top = tk.Tk()

top.title('Sqiarepf a number')
top.geometry('700x500')

data = tk.IntVar()

def square():
    num = data.get()
    result = num*num
    messagebox.showinfo('MESSAGE','Square is %i' %result)
    
    

l1 =tk.Label(top,text='INPUT').place(x=50,y=50)
e1 =tk.Entry(top,textvariable = data).place(x=100,y=50)
b1 =tk.Button(top,text = 'Square',command = square,font=('bold',14),bg='purple',fg='yellow').place(x=150,y=100)

top.mainloop()

```

```python
#take 2 numbers,add,sub,divide,mul

top = tk.Tk()

top.title('Arithmatic operations')
top.geometry('700x500')
data1 = tk.IntVar()
data2 = tk.IntVar()

def addition():
    num1 = data1.get()
    num2 = data2.get()
    result = num1+num2
    messagebox.showinfo('MESSAGE','Addition is %i' %result)

def Substraction():
    num1 = data1.get()
    num2 = data2.get()
    result = num1-num2
    messagebox.showinfo('MESSAGE','Substraction is %i' %result)

def Multiplication():
    num1 = data1.get()
    num2 = data2.get()
    result = num1*num2
    messagebox.showinfo('MESSAGE','Multiplication is %i' %result)

def Division():
    num1 = data1.get()
    num2 = data2.get()
    result = num1/num2
    messagebox.showinfo('MESSAGE','Division is %i' %result)

l1 =tk.Label(top,text='INPUT1').place(x=50,y=50)
e1 =tk.Entry(top,textvariable = data1).place(x=100,y=50)
l2 =tk.Label(top,text='INPUT2').place(x=50,y=100)
e2 =tk.Entry(top,textvariable = data2).place(x=100,y=100)

b1 =tk.Button(top,text = 'addition',command = addition,font=('bold',14),bg='red',fg='yellow').place(x=200,y=150)  #for addition button

b2 =tk.Button(top,text = 'Substraction',command = Substraction,font=('bold',14),bg='red',fg='yellow').place(x=300,y=150)   #for substraction button


b3 =tk.Button(top,text = 'Multiplication',command = Multiplication,font=('bold',14),bg='red',fg='yellow').place(x=440,y=150) #for multiplication button


b4 =tk.Button(top,text = 'Division',command = Division,font=('bold',14),bg='red',fg='yellow').place(x=570,y=150) #for division button



top.mainloop()


```
